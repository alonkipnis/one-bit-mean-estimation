```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('ggplot')
%matplotlib inline
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from scipy.stats import norm as norm

fig = plt.figure()
ax = fig.gca(projection='3d')

r = np.random.randn(1000,2)
H, edges = np.histogramdd(r)

# Make data.
X = (edges[0][:-1] + edges[0][1:])/2
Y = (edges[1][:-1] + edges[1][1:])/2
X, Y = np.meshgrid(X, Y)
Z = H

#true_cumH = norm.pdf(X)*norm.cdf(Y)

H = H / np.sum(H)
cumH = H.cumsum(1).cumsum(0)

true_H = norm.pdf(Y)*norm.pdf(X)
true_H = true_H / np.sum(true_H)

true_cumH = true_H.cumsum(1).cumsum(0)

# Plot the surface.
#surf = ax.plot_surface(X, Y, Z,linewidth=0, antialiased=False)
diff = cumH - true_cumH 
#diff = H / np.sum(H)- true_H / np.sum(true_H)
surf = ax.plot_surface(X, Y, diff,linewidth=0, antialiased=False)

plt.show()
plt.imshow(diff) 
plt.show()
```
